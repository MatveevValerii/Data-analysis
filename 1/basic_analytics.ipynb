```python
import pandas as pd
import matplotlib.pyplot as plt 
import csv 
import seaborn as sns

df = pd.read_csv(r'datafile1.csv')
df['var2'] = [x if (x < 1000)&(x>-1000) else None for x in df['var2']]
df['var1'] = [x if (x < 1000)&(x>-1000) else None for x in df['var1']]
df['var3'] = [x if (x < 1000)&(x>-1000) else None for x in df['var3']]
df = df.dropna()
print(df)
```

```python
sns.boxplot(df['var1'])
plt.show()
```

```python
sns.boxplot(df['var2'])
plt.show()
```

```python
sns.boxplot(df['var3'])
plt.show()
```

```python
sns.histplot(df['var1'])
plt.show()
sns.histplot(df['var2'])
plt.show()
sns.histplot(df['var3'])
plt.show()
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
xs = df['var1']
ys = df['var2']
zs = df['var3']
ax.scatter(xs, ys, zs)

ax.set_xlabel('variable 1')
ax.set_ylabel('variable 2')
ax.set_zlabel('variable 3')

plt.show()
```

```python
sns.heatmap(df.corr())
plt.show()
```
